```python
import main as mc
import numpy as np
import matplotlib.pyplot as plt

x = 10 * (np.random.rand(100) - 0.5)
y = x + np.sin(x) + np.cos(x)

m1 = mc.MCRegressor(hiddenCounts=[25, 25], activations=['lin', 'relu', 'sig', 'lin'])
m2 = mc.MCRegressor(hiddenCounts=[25, 25], activations=['lin', 'relu', 'tanh', 'lin'])
m3 = mc.MCRegressor(hiddenCounts=[25, 25], activations=['lin', 'silu', 'dsilu', 'lin'])

m1.fit(x, y, Ieta=15)
m2.fit(x, y, Ieta=15)
m3.fit(x, y, Ieta=15)

plt.plot(y)
plt.plot(m1.predict(x))
plt.plot(m2.predict(x))
plt.plot(m3.predict(x))
plt.legend(["True", "sig", "tanh", "silu"])
plt.show()
```
